# Binary Classification

### Import

```python
import os
import cv2
import shutil
from matplotlib import pyplot as plt
```

```python
object_to_detect = "fiducial"

resized_images = "./data/resized/"
labels_folder = "BBox_Label_Tool/Labels/001/"
SHOW_IMAGES = False
```

### Link each image to its bounding boxes label

```python
rects = {}
for file in os.listdir(labels_folder):
    with open(labels_folder + file, "r") as text_file: 
        lines = text_file.readlines()[1:]
        content = [l.replace("\n", "") for l in lines]
        labels = set([rect.split(" ")[-1] for rect in content])
        tmp = {}
        for lab in labels:
            tmp[lab] = [rect.split(" ")[:-1] for rect in content if lab in rect]
        rects[file[:-4]] =  tmp

data = {}
for img in os.listdir(resized_images):
    if img[-4:] == ".jpg":
        ID = img[:-4]
        image = cv2.imread(resized_images + img,1)

        data[ID] = {"image": image, "rects": rects[ID]}
```

### Create the file tree

```python
if os.path.exists('/data/binary_classification/'):
    shutil.rmtree('/data/binary_classification/')
if not os.path.exists("./callbacks/"):
    os.makedirs("./callbacks/")
if not os.path.exists("./data/binary_classification/"+str(object_to_detect)+"/train/with/"):
    os.makedirs("./data/binary_classification/"+str(object_to_detect)+"/train/with/")
if not os.path.exists("./data/binary_classification/"+str(object_to_detect)+"/train/without/"):
    os.makedirs("./data/binary_classification/"+str(object_to_detect)+"/train/without/")
if not os.path.exists("./data/binary_classification/"+str(object_to_detect)+"/test/with/"):
    os.makedirs("./data/binary_classification/"+str(object_to_detect)+"/test/with/")
if not os.path.exists("./data/binary_classification/"+str(object_to_detect)+"/test/without/"):
    os.makedirs("./data/binary_classification/"+str(object_to_detect)+"/test/without/")
```

### Separate the pictures containing the "object_to_detect" from the other

```python
with_list = []
without_list = []
for elem in data:
    if object_to_detect in data[elem]["rects"].keys():
        with_list.append([elem, data[elem]["image"]])
    else:
        without_list.append([elem, data[elem]["image"]])
```

### Separate the data in train/test sets

```python
train_with = with_list[:int(0.8*len(with_list))]
test_with = with_list[int(0.8*len(with_list)):]
train_without = without_list[:int(0.8*len(without_list))]
test_without = without_list[int(0.8*len(without_list)):]
```

```python
for image in train_with:
    image[1] = cv2.resize(image[1],(150,150))
    cv2.imwrite("./data/binary_classification/"+str(object_to_detect)+"/train/with/" +str(image[0])+".jpg", image[1])
for image in test_with:
    image[1] = cv2.resize(image[1],(150,150))
    cv2.imwrite("./data/binary_classification/"+str(object_to_detect)+"/test/with/" +str(image[0])+".jpg", image[1])
for image in train_without:
    image[1] = cv2.resize(image[1],(150,150))
    cv2.imwrite("./data/binary_classification/"+str(object_to_detect)+"/train/without/" +str(image[0])+".jpg", image[1])
for image in test_without:
    image[1] = cv2.resize(image[1],(150,150))
    cv2.imwrite("./data/binary_classification/"+str(object_to_detect)+"/test/without/" +str(image[0])+".jpg", image[1])
```

# The model

### Import

```python
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Model
from keras.layers import Flatten, Dense, Dropout, Activation, Conv2D, MaxPooling2D
from keras.applications.vgg16 import VGG16
from keras.callbacks import ModelCheckpoint
```

### The model

```python
vgg_model = VGG16(weights='imagenet',
                               include_top=False,
                               input_shape=(150, 150, 3))

# Stacking a new simple convolutional network on top of it    
x = Conv2D(filters=64, kernel_size=(3, 3), activation='relu')(vgg_model.output)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Flatten()(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(1, activation='sigmoid')(x)

model = Model(input=vgg_model.input, output=x)

# Make sure that the pre-trained bottom layers are not trainable
for layer in model.layers[:15]:
    layer.trainable = False
    
model.compile(loss='binary_crossentropy',
                     optimizer='adam',
                     metrics=['accuracy'])
```

### The data generator

```python
batch_size = 128

# this is the augmentation configuration we will use for training
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True)

# this is the augmentation configuration we will use for testing:
test_datagen = ImageDataGenerator(rescale=1./255)

# this is a generator that will read pictures found in
# subfolers of 'data/train', and indefinitely generate
# batches of augmented image data
train_generator = train_datagen.flow_from_directory(
        "./data/binary_classification/"+str(object_to_detect)+"/train/",
        target_size=(150, 150),  # all images will be resized to 150x150
        batch_size=batch_size,
        class_mode='binary') 

# this is a similar generator, for validation data
validation_generator = test_datagen.flow_from_directory(
        "./data/binary_classification/"+str(object_to_detect)+"/test/",
        target_size=(150, 150),
        batch_size=batch_size,
        class_mode='binary')
```

### Train the model

```python

chk = ModelCheckpoint("./callbacks/weights.{epoch:02d}-{val_loss:.2f}.hdf5", monitor='val_loss', verbose=0, save_best_only=True, save_weights_only=False, mode='auto', period=1)

history = model.fit_generator(
        train_generator,
        steps_per_epoch=500 // batch_size,
        epochs=1,
        callbacks=[chk],
        validation_data=validation_generator,
        validation_steps=200 // batch_size)
```

### Plot the training history

```python
#get the details form the history object
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

#Train and validation accuracy
plt.plot(epochs, acc, 'b', label='Training accurarcy')
plt.plot(epochs, val_acc, 'r', label='Validation accurarcy')
plt.title('Training and Validation accurarcy')
plt.legend()

plt.figure()
#Train and validation loss
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and Validation loss')
plt.legend()
plt.show()
```

### Go back to the best step

```python
best_loss = 1
best_weights = None
for w in os.listdir("./callbacks/"):
    val_loss = float(w.split("-")[1].split(".hdf5")[0])
    if val_loss < best_loss:
        best_loss = val_loss
        best_weights = w
print(best_weights)
model.load_weights("./callbacks/" +best_weights)
```

### Show example on test data

```python
test_generator = test_datagen.flow_from_directory(
        "./data/binary_classification/"+str(object_to_detect)+"/test/",
        target_size=(150, 150),
        batch_size=1,
        class_mode='binary')

i = 0
columns = 5
text_labels = []
plt.figure(figsize=(30,20))
for batch in test_generator:
    img = batch[0]
    pred = model.predict_on_batch(img)[[0]]
    if pred > 0.5:
        text_labels.append("With ")
    elif pred <= 0.5:
        text_labels.append("Without ")
    else:
        text_labels.append("ERROR")
    plt.subplot(5 / columns + 1, columns, i + 1)
    plt.title(text_labels[i] + str(object_to_detect))
    imgplot = plt.imshow(img[0])
    i += 1
    if i % 10 == 0:
        break
plt.show()
```
